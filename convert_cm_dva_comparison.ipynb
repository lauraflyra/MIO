# Plotting different conversions from cm to dva

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cm2dva(size_cm, dist_cm):
    """
    size_cm:  size of an object in the screen in centimeters
    dist_cm: distance from the participant to the screen in centimeter

    return: size of the object in degrees visual angle
    """
    size_dva = 2*np.arctan2(size_cm,(2*dist_cm))*180/np.pi
    return size_dva
```

```python
#First method: use arctan formula to get cm2dva for all different sizes in cm
dist_to_screen = 22.5 #cm
cm_array = np.linspace(0,100,100)
cm2dva_array = cm2dva(cm_array,dist_to_screen)
```

```python
#Second method: get the size of 1 cm in dva and then scale it
cm2dva_unit = cm2dva(1,dist_to_screen)
cm2dva_unit_array = cm2dva_unit*cm_array
plt.plot(cm_array, cm2dva_unit_array, label='scaling')
plt.plot(cm_array,cm2dva_array, label='tangent')
plt.xlabel("cm")
plt.ylabel("dva")
plt.title("Conversion from cm to dva using viewing distance of 22.5 cm")
plt.legend()
plt.show()
```
